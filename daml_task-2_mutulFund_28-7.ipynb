```python
import requests
url=requests.get("https://api.mfapi.in/mf")
```

```python
mydata = url.json()
```

```python
print(type(mydata))
```

## Print total count of mutual fund

```python
print("Total number of days:",len(mydata))
```

## Print all data using for loop (scheme code:—— scheme name: ———)

```python
for i in range(0,len(mydata)):
    print("Scheme Code:",mydata[i]["schemeCode"],
          "Scheme Name:",mydata[i]["schemeName"])
```

## Allow user to insert scheme code → search → print scheme name else no scheme found)

```python
userDate = input("Enter scheme code: ")

found_data = False
for item in mydata:
    if userDate == str(item["schemeCode"]):
        print("Scheme Code:", item["schemeCode"])
        print("Scheme Name:", item["schemeName"])
        found_data = True
        break

if not found_data:
    print("Scheme name not found for scheme code:", userDate)
```
